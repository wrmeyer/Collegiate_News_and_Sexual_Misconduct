```python
import pandas as pd
import os
location = str(os.getcwd())
```

```python
f = (location[:-8] + "Articles\\articles_cleaned.csv")
df= pd.read_csv(f)
df = df.to_dict("index")
```

```python
#cleaning body text
for i in range(len(df)):
    df[i]["body"] = str(df[i]["body"]).replace("\n", "").replace("\r", "").replace(".", ". ")
df[10000]["body"]
```
